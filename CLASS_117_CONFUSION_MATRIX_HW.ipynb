# ***`IMPORT's`***

```python
import pandas as pd
import csv
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
```

# ***`GOING FOR THE DATAFRAME`***

```python
df = pd.read_csv("BankNote_Authentication.csv")
print(df.head())
```

# ***`TRAINING AND TESTING THE DATA`***

```python
variance = df['variance']
class_df = df['class']

variance_train,variance_test,class_df_train,class_df_test = train_test_split(variance , class_df,test_size = 0.25,random_state = 0)
```

# ***`TRAINING WITH LOGISTIC REGRESSION`***

```python
X_train = np.reshape(variance_train.ravel(),(len(variance_train),1))
Y_train = np.reshape(class_df_train.ravel(),(len(class_df_train),1))

classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train,Y_train)
print("X TRAIN IS : \n\n\n",X_train,"\n\n\nY TRAIN : \n\n\n",Y_train,"\n\n CLASSIFIER : \n\n",classifier,"\n\n FITTED CLASSIFIER : \n\n",classifier.fit)
```

# ***`TESTING WITH LOGISTIC REGRESSION`***

```python
X_test = np.reshape(variance_test.ravel(),(len(variance_test),1))
Y_test = np.reshape(class_df_test.ravel(),(len(class_df_test),1))

class_df_predict = classifier.predict(X_test)
print("CLASS OF DATAFRAME PREDICTED :\n\n",class_df_predict,"\n\n\nX TEST IS:\n\n\n",X_test,"\n\n\nY TEST IS:\n\n\n",Y_test)
```

# ***`PREDICTING VALUES`***

```python
predicted_val = []

for i in class_df_predict:
  if i == 0:
    predicted_val.append("No")
  else :
    predicted_val.append("Yes")
print("VALUES HAVE BEEN PREDICTED IN THE DATA : \n\n",predicted_val)
```

# ***`ACTUAL VALUES`***

```python
actual_val = []

for i in Y_test.ravel():
  if i == 0 :
    actual_val.append("No")
  else :
    actual_val.append("Yes")
print("ACTUAL VALUES IN THE DATA : \n\n",actual_val)
```

# ***`CONFUSION MATRIX GRAPH`***

```python
labels = ["Yes","No"]

cm = confusion_matrix(actual_val,predicted_val,labels)

a_x = plt.subplot()
sns.heatmap(cm, annot = True, ax = a_x)
a_x.set_xlabel("Predicted Value")
a_x.set_ylabel("Actual Value")
a_x.set_title("Confusion Matrix")
a_x.xaxis.set_ticklabels(labels);
a_x.yaxis.set_ticklabels(labels);
```
